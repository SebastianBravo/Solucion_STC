###**2.** Graficar la parte par e impar de las siguientes funciones: 

####**a.** $a(t) = 
\begin{cases} 
  t & \text{$|t|<4$} \\ 
  0 & \text{$|t|\geq4$}
\end{cases}$

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector tiempo
t = np.linspace(-6,6,1000)
a = []
a_par = []
a_impar =[]

# Evalua cada tiempo en la señal
for x in t:
  if (abs(x)<4):
    a.append(x)
  else:
    a.append(0)

# Invierte la señal en tiempo
a_inv = list(reversed(a))

# Calcula la parte par e impar de la señal
for i in range(len(a)):
  a_par.append((a[i] + a_inv[i])/2)
  a_impar.append((a[i] - a_inv[i])/2)

# Gráfica de la señal
plt.plot(t,a)
plt.xlabel('t')
plt.ylabel('a(t)')
plt.title('Señal a(t)')
plt.grid(True)
plt.show()

# Gráfica de la componente par de la señal
plt.plot(t,a_par, label='Parte par')
plt.xlabel('t')
plt.title('Componente par de a(t)')
plt.ylim(-4, 4)  
plt.grid(True)
plt.show()

# Gráfica de la componente impar de la señal
plt.plot(t,a_impar, label='Parte impar')
plt.xlabel('t')
plt.title('Componente impar de a(t)')
plt.grid(True)
plt.show()
```
