##**Taller No. 15** Bases ortogonales, bases ortonormales de espacios lineales y expresión de elementos.

###**1.** Sea V el espacio lineal definido como el conjunto de elementos definidos como:

$$A\ =\ \begin{bmatrix}
a_{11} & a_{12}\\
a_{21} & a_{22}
\end{bmatrix}$$

### y su producto interior de **V** se define como:

$$\langle A,B\rangle =\frac{1}{4} \cdot \sum _{n=1}^{2}\sum _{n=1}^{2} a_{nm} \cdot b_{nm}$$

####**a.** El siguiente conjunto de elementos: ¿$X_1$, $X_2$, $X_3$, y $X_4$ forman una base de **V**?

$$X_{1} \ =\ \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix} ,\ X_{2} \ =\ \begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix} ,\ X_{3} \ =\ \begin{bmatrix}
1 & -1\\
1 & -1
\end{bmatrix} ,\ X_{4} \ =\ \begin{bmatrix}
2 & 2\\
0 & 0
\end{bmatrix}$$

Se comprueba ortogonalidad:

 \begin{array}{l}
\langle X_{1} ,X_{2} \rangle =\frac{1}{4} \cdot ( 0+0+0+0) \ =\ 0\\
\langle X_{1} ,X_{3} \rangle =\frac{1}{4} \cdot ( 1+0+0-1) \ =\ 0\\
\langle X_{1} ,X_{4} \rangle =\frac{1}{4} \cdot ( 2+0+0+0) \ =\ \frac{1}{2}\\
\langle X_{2} ,X_{3} \rangle =\frac{1}{4} \cdot ( 0-1+1+0) \ =\ 0\\
\langle X_{2} ,X_{4} \rangle =\frac{1}{4} \cdot ( 0+2+0+0) \ =\ \frac{1}{2}\\
\langle X_{3} ,X_{4} \rangle =\frac{1}{4} \cdot ( 2-2+0+0) \ =\ 0
\end{array}

No forma una base ya que no se cumple ortogonalidad

####**b**. A partir de los elementos $X_1$, $X_2$, $X_3$, y $X_4$, encontrar una base ortogonal ($Y_1$, $Y_2$, $Y_3$, y $Y_4$) para **V** 

Siguiendo el método de Gram-Schmmidt:

Para $\displaystyle Y_{2}$


\begin{gather*}
Y_{1} =X_{1} =\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}\\
\\
\langle Y_{1} ,Y_{1} \rangle =\frac{1}{4} \cdot ( 1+0+0+1) \ =\ \frac{1}{2}\\
\\
Y_{2} =X_{2} -\sum _{i=1}^{1}\frac{\langle X_{2} ,Y_{i} \rangle }{\langle Y_{i} ,Y_{i} \rangle } \cdot Y_{i} =X_{2} -\frac{0}{\frac{1}{2}} Y_{1} =X_{2} =\ \begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}\\
\end{gather*}
Para $\displaystyle Y_{3}$


\begin{gather*}
\langle Y_{2} ,Y_{2} \rangle =\frac{1}{4} \cdot ( 0+1+1+0) \ =\ \frac{1}{2}\\
\\
Y_{3} =X_{3} -\sum _{i=1}^{2}\frac{\langle X_{3} ,Y_{i} \rangle }{\langle Y_{i} ,Y_{i} \rangle } \cdot Y_{i} =X_{3} -\left(\frac{0}{\frac{1}{2}} Y_{1} +\frac{0}{\frac{1}{2}} Y_{2}\right) =X_{3} =\begin{bmatrix}
1 & -1\\
1 & -1
\end{bmatrix}
\end{gather*}


Para $\displaystyle Y_{4}$


\begin{gather*}
\langle Y_{3} ,Y_{3} \rangle =\frac{1}{4} \cdot ( 1+1+1+1) \ =\ 1\\
\\
Y_{3} =X_{4} -\sum _{i=1}^{3}\frac{\langle X_{4} ,Y_{i} \rangle }{\langle Y_{i} ,Y_{i} \rangle } \cdot Y_{i} =X_{4} -\left(\frac{\frac{1}{2}}{\frac{1}{2}} Y_{1} +\frac{\frac{1}{2}}{\frac{1}{2}} Y_{2} +\frac{0}{1} Y_{3}\right)\\
\\
Y_{3} \ =\ X_{4} -Y_{1} -Y_{2} =\begin{bmatrix}
2 & 2\\
0 & 0
\end{bmatrix} -\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix} -\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix} =\begin{bmatrix}
1 & 1\\
-1 & -1
\end{bmatrix}\\
\end{gather*}
La base ortogonal resultante es:


\begin{equation*}
Y_{1} \ =\ \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix} ,\ Y_{2} \ =\ \begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix} ,\ Y_{3} \ =\ \begin{bmatrix}
1 & -1\\
1 & -1
\end{bmatrix} ,\ Y_{4} \ =\ \begin{bmatrix}
1 & 1\\
-1 & -1
\end{bmatrix}
\end{equation*}

####**c.** Ortonormalizar la base ortogonal: 




Cálculo de normas:


\begin{gather*}
\sqrt{\langle Y_{1} ,Y_{1} \rangle } =\sqrt{\frac{1}{2}} \ \\
\\
\sqrt{\langle Y_{2} ,Y_{2} \rangle } =\sqrt{\frac{1}{2}} \ \\
\\
\sqrt{\langle Y_{3} ,Y_{3} \rangle } =\sqrt{1} =1\ \\
\\
\sqrt{\langle Y_{4} ,Y_{4} \rangle } =\sqrt{1} =1\ 
\end{gather*}


Dividiendo los elementos entre sus normas:


\begin{equation*}
\widehat{Y_{1}} \ =\ \begin{bmatrix}
\sqrt{2} & 0\\
0 & \sqrt{2}
\end{bmatrix} ,\ \widehat{Y_{2}} \ =\ \begin{bmatrix}
0 & \sqrt{2}\\
\sqrt{2} & 0
\end{bmatrix} ,\ \widehat{Y_{3}} \ =\ \begin{bmatrix}
1 & -1\\
1 & -1
\end{bmatrix} ,\ \widehat{Y_{4}} \ =\ \begin{bmatrix}
1 & 1\\
-1 & -1
\end{bmatrix}
\end{equation*}

####**d.** Exprese el elemento $\displaystyle Z=\begin{bmatrix}
3 & 1\\
3 & 5
\end{bmatrix}$ , en función de los elementos de la base ortonormal hallada en el literal c, es decir, $\displaystyle Z\ =\ c_{1}\widehat{Y_{1}} +c_{2}\widehat{Y_{2}} +c_{3}\widehat{Y_{3}} +c_{4}\widehat{Y_{4}}$

 \begin{array}{l}
c_{1} =\langle Z,\widehat{Y_{1}} \rangle =\frac{1}{4} \cdot \left( 3\sqrt{2} +5\sqrt{2}\right) =2\sqrt{2}\\
\\
c_{2} =\langle Z,\widehat{Y_{2}} \rangle =\frac{1}{4} \cdot \left( 1\sqrt{2} +3\sqrt{2}\right) =\sqrt{2}\\
\\
c_{3} =\langle Z,\widehat{Y_{3}} \rangle =\frac{1}{4} \cdot ( 3-1+3-5) =0\\
\\
c_{4} =\langle Z,\widehat{Y_{4}} \rangle =\frac{1}{4} \cdot ( 3+1-3-5) =-1\\
\\
Z\ =\ 2\sqrt{2} \cdot \widehat{Y_{1}} +\sqrt{2} \cdot \widehat{Y_{2}} -\widehat{Y_{4}}
\end{array}

###**6.** Se tiene la siguiente base ortonormal de duración un (1) segundo, mostrada en la siguiente figura:

Dicha base ortonormal tiene una equivalencia para la trasmisión de datos, así:

$$x ̂_1 (t)=O, x ̂_2 (t)=M,x ̂_3 (t)=L, x ̂_4 (t)=A.$$

De tal modo que al realizar una combinación lineal con los elementos de la base se obtiene una señal que pertenece al espacio y equivale a un mensaje (secuencia de letras), siguiendo las siguientes regla:


> *La señal de la base ortonormal con mayor amplitud corresponderá a la primera letra de la palabra enviada,  la segunda señal con mayor amplitud corresponderá al segunda letra y así sucesivamente.*

De esta forma, tomando por ejemplo la señal:

$$z(t)=1⋅x ̂_1 (t)+ 4⋅x ̂_2 (t)+2⋅x ̂_3 (t)+3⋅x ̂_4 (t).$$


mostrada en la siguiente figura:

taller15-6-2.png

Equivaldría al siguiente mensaje: **MALO**, debido a que la mayor amplitud en la señal la tiene el elemento $x ̂_2 (t)=M$, la segunda amplitud corresponde al elemento $x ̂_4 (t)=A$, la tercera amplitud corresponde al elemento $x ̂_3 (t)=L$ y la amplitud más baja corresponde al elemento $x ̂_1 (t)=O$.

El producto interior se define como:

$$\langle x_{1}( t) ,x_{2}( t) \rangle =\int _{0}^{4} \ x_{1}( t) \cdot x_{2}( t) dt$$
