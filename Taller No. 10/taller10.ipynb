####**f.**	El conjunto de señales sinusoidales con periodo T=2, asociadas al campo escalar real

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = 0

# Tiempo final de la gráfica
t_final = 10

# Número de puntos a graficar
puntos = 10000

# Frecuencia angular
w = 2*np.pi*0.5

# Fases
fase1 = 0
fase2 = np.pi/3
fase3 = 2*np.pi/3

# Amplitudes
A1 = 4
A2 = 2
A3 = 3

# No modificar a partir de este punto
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Para cada valor en t, calcula la señal
sig1 = A1*np.sin(w*t - fase1)
sig2 = A2*np.sin(w*t - fase2)
sig3 = A3*np.sin(w*t - fase3)

# Gráfica de las señales
plt.plot(t,sig1, label = r"$x_1(t)$", linestyle='-.')
plt.plot(t,sig2, label = r"$x_2(t)$", linestyle='dotted')
plt.plot(t,sig3, label = r"$x_3(t)$", linestyle='solid')
plt.xlabel('t')
plt.ylabel(r'Amplitud')
plt.title(r'Señales con periodo T=2')
plt.legend()
plt.grid(True)
plt.show()
```

####**i.** El conjunto de señales a trozos definidas en el intervalo [0,4), con valor constante de duración un segundo por trozo, asociadas al campo escalar real.

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = 0

# Tiempo final de la gráfica
t_final = 4

# Número de puntos a graficar
puntos = 10000

# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Cálculo de señales
sig1 = []
sig2 = []
sig3 = []

# Señal 1
for x in t:
  if (x>=0) and (x<=1):
    sig1.append(1)
  else:
    sig1.append(0)

# Señal 2
for x in t:
  if (x>=2) and (x<=3):
    sig2.append(2)
  else:
    sig2.append(0)

# Señal 3
for x in t:
  if (x>=3) and (x<=4):
    sig3.append(3)
  else:
    sig3.append(0)

# Gráfica de las señales
plt.plot(t,sig1, label = r"$x_1(t)$", linestyle='solid')
plt.plot(t,sig2, label = r"$x_2(t)$", linestyle='-.')
plt.plot(t,sig3, label = r"$x_3(t)$", linestyle='dotted')
plt.xlabel('t')
plt.ylabel(r'Amplitud')
plt.title(r'Señales con periodo T=2')
plt.legend()
plt.grid(True)
plt.show()
```
