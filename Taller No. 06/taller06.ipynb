###**1.** Graficar las siguientes señales. Indique tiempos y amplitudes relevantes:

####**a.** $a\left(t\right)=4\delta\left(t\right)\cdot \cos\left(\frac{5\pi}{4}t\right)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector tiempo
t = np.linspace(-2,2,1000000)

def dirac_delta(t, a, b, t0, c):
  """
  Función que crea señal au(bt + t0) + c 
  Toma como parametros un vector de tiempo(t) y a, b, t0 y c
  Retorna una pareja de tiempo y amplitud
  """
  # Entre más pequeño sea delta, mejor es la representación del impulso
  delta = 1/1000
  amp = []
  
  # Crea funcion dirac_delta(t)
  for x in t:
    if (x >= -delta/2) and (x <= delta/2):
      amp.append(1/delta)
    else:
      amp.append(0)

  # Desplazamiento y escalado en tiempo
  t = [(x+t0)/b for x in t]

  # Desplazamiento en eje vertical y escalado en amplitud
  amp = [(i*a)+c for i in amp]

  return (t, amp)

def integral(t, amp):
  """
  Función que calcula la integral de una señal
  Toma como parametros una pareja de (tiempo, amplitud)
  Retorna el resultado de la integral
  """
  dt = (max(t)- min(t))/len(t)
  area = 0

  for a in amp:
    area += dt*a

  return area

# Define la señal
t, delta = dirac_delta(t, 4, 1, 0, 0)
cos = [np.cos((5*np.pi/4)*x) for x in t]
sig = []

for i in range(len(t)):
  sig.append(cos[i]*delta[i])

# Imprime el area bajo la curva de la señal
print(f"El área bajo la curva de la señal es: {integral(t, sig)}")

# Gráfica de la señal
plt.plot(t,sig)
plt.xlabel('t')
plt.ylabel(r'$a\left(t\right)=4\delta\left(t\right)\cdot \cos\left(\frac{5\pi}{4}t\right)$')
plt.title(r'Señal $a\left(t\right)=4\delta\left(t\right)\cdot \cos\left(\frac{5\pi}{4}t\right)$')
plt.grid(True)
plt.show()
```

###**2.** Graficar la siguiente señal $y\left(4-t\right)$, sabiendo que:

$y\left(t\right)=3e^{-2t}\cdot u\left(t-1\right)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -2

# Tiempo final de la gráfica
t_final = 6

# Número de puntos a graficar
puntos = 10000

# No modificar
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# No modificar
# Para cada valor del vector tiempo, calcula la señal con transformaciones
y = []
for i in t:
  if i>=1:
    y.append(3*np.exp(-2*i))
  else:
    y.append(0)

# Transformaciones a realizar: ax(bt - t0) + c
a = 1
b = -1
t0 = -4
c = 0

# A partir de este punto, no modificar
# Desplazamiento y escalado en tiempo
t_mod = [(i+t0)/b for i in t]

# Desplazamiento en eje vertical y escalado en amplitud
y_mod = [(i*a)+c for i in y]

# Gráfica de la señal sin transformaciones
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel(r'$y(t)$')
plt.title(r'Señal $y(t)$')
plt.grid(True)
plt.show()

# Gráfica de la señal con transformaciones
plt.plot(t_mod,y_mod)
plt.xlabel('t')
plt.ylabel(r'$y(4-t)$')
plt.title(r'Señal $y(4-t)$')
plt.grid(True)
plt.show()
```
