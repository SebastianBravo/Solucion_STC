###Graficar:

####**a.** $x({3t + \frac{1}{3}})$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -4

# Tiempo final de la gráfica
t_final = 4

# Número de puntos a graficar
puntos = 10000

# No modificar
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# No modificar
# Para cada valor del vector tiempo, calcula x
x = []
for i in t:
  if (i>=-2) and (i<=-1):
    x.append((2*i)+ 4)
  elif (i>-1) and (i<=1):
    x.append(2)
  elif (i>1) and (i<=3):
    x.append(-i + 3)
  else:
    x.append(0)

# Transformaciones a realizar: ax(bt - t0) + c
a = 1
b = 3
t0 = -1/3
c = 0

# A partir de este punto, no modificar
# Desplazamiento y escalado en tiempo
t_mod = [(i+t0)/b for i in t]

# Desplazamiento en eje vertical y escalado en amplitud
x_mod = [(i*a)+c for i in x]

# Gráfica de la señal
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel(r'$x(t)$')
plt.title(r'Señal $x(t)$')
plt.grid(True)
plt.show()

# Gráfica de la señal modificada
t0_mod = -t0 if t0 < 0 else t0
sign = '+' if t0 < 0 else '-'
nombre_x_mod = f"${a}x({b}t {sign} {'%.2f' % t0_mod}) + {c}$"

plt.plot(t_mod,x_mod)
plt.xlabel('t')
plt.ylabel(nombre_x_mod)
plt.title(f"señal {nombre_x_mod}")
plt.grid(True)
plt.show()
```

#### Graficar la señal $x(-3t-6)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -12

# Tiempo final de la gráfica
t_final = 14

# Número de puntos a graficar
puntos = 10000

# No modificar
# Vector tiempo
t_ini = np.linspace(t_inicial,t_final,puntos)

# No modificar
# Para cada valor del vector tiempo, calcula la señal con transformaciones
x_ini = []
for i in t_ini:
  if (i>=-6) and (i<=0):
    x_ini.append(2)
  elif (i>0) and (i<=12):
    x_ini.append(4)
  else:
    x_ini.append(0)

# No modificar
# Calcula la señal x(t)
x = [i for i in x_ini]
t = [i/2 - 3 for i in t_ini]

# Transformaciones a realizar: ax(bt - t0) + c
a = 1
b = -3
t0 = 6
c = 0

# A partir de este punto, no modificar
# Desplazamiento y escalado en tiempo
t_mod = [(i+t0)/b for i in t]

# Desplazamiento en eje vertical y escalado en amplitud
x_mod = [(i*a)+c for i in x]

# Gráfica de la señal inicial
plt.plot(t_ini,x_ini)
plt.xlabel('t')
plt.ylabel(r'$x(\frac{t-6}{2})$')
plt.title(r'Señal $x(\frac{t-6}{2})$')
plt.grid(True)
plt.show()

# Gráfica de la señal sin transformaciones
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel(r'$x(t)$')
plt.title(r'Señal $x(t)$')
plt.grid(True)
plt.show()

# Gráfica de la señal con transformaciones
plt.plot(t_mod,x_mod)
plt.xlabel('t')
plt.ylabel(r'$x(-3t-6)$')
plt.title(r'Señal $x(-3t-6)$')
plt.grid(True)
plt.show()
```
