###**6.** Se tiene la siguiente relación entrada-salida de un SLIT:

$$y( t) =3x( t-1)$$

####**a.** Hallar la respuesta impulso del sistema:



$$h( t) \ =\ 3\delta ( t-1)$$

####**b.** Hallar la salida del sistema si la entrada fuera:

$$x_{1}( t) =-2\delta ( t+1) +2\delta ( t-1)$$


$$y_{1}( t) =6[ -\delta ( t) +\delta ( t-2)]$$

###**7.** Se tiene la siguiente relación entrada-salida de un SLIT:

$$y( t) =\int _{t-3}^{t-1} x( \tau ) d\tau $$

####**a.** Hallar la respuesta impulso del sistema:



$$h( t) =\int _{t-3}^{t-1} \delta ( \tau ) d\tau =u( t-1) -u( t-3)$$

####**b.** Evaluar el valor de la respuesta impulso $h(t)$ del sistema en $t=2$



$$h( 2) \ =\ 1$$

###**8.** Se tiene un SLIT con la siguiente respuesta impulso:

$$h( t) \ =\ 3u( t-1) -3u( t-3)$$

####**a.** Hallar la salida del sistema si la entrada fuera:

$$x_{1}( t) =2\delta ( t) +2\delta ( t-2)$$


$$y_{1}( t) =\ 6[ u( t-1) -u( t-3)] +6[ u( t-3) -3u( t-5)]$$

####**b.** Hallar la salida del sistema si la entrada fuera:

$$x_{2}( t) =-\delta ( t) +\delta ( t-2) -\delta ( t-4)$$


$$y_{2}( t) =\ -3[ u( t-1) -u( t-3)] +3[ u( t-3) -3u( t-5)] -3[ u( t-5) -3u( t-7)]$$
