####**c.** Dado el siguiente conjunto de elementos, indicar cuales de ellos son ortogonales a otros elementos del mismo conjunto, utilizando el productor interior definido como: 

$$⟨A,B⟩=a_1\cdot b_1+a_2\cdot b_2$$

|||
| --- | --- |
| $\displaystyle A\ =\ ( 1,0)$ |$\displaystyle E\ =\ \left(\frac{\sqrt{2}}{2} ,\frac{\sqrt{2}}{2}\right)$|
|$\displaystyle B\ =\ ( 0,1)$|$F =(\sqrt{2} ,-\sqrt{2})$|
|$\displaystyle C\ =\ ( 1,1)$|$\displaystyle G\ =\ ( 1,2)$|
|$\displaystyle D\ =\ ( 1,-1)$|$\displaystyle H\ =\ ( 4,-2)$|

```python
import numpy as np

# Función para calcular productor interior
def prod_interior(a, b):
    result = a[0]*b[0] + a[1]*b[1]
    return result

# Conjunto de datos:
A = [1,0]
B = [0,1]
C = [1,1]
D = [1,-1]
E = [np.sqrt(2)/2, np.sqrt(2)/2]
F = [np.sqrt(2), -np.sqrt(2)]
G = [1,2]
H = [4,-2]

conjunto = [A,B,C,D,E,F,G,H]
etiquetas = ['A','B','C','D','E','F','G','H']

# Prueba de ortogonalidad:
# Para cada par de elementos del conjunto, calcula el producto interior y lo imprime
for i in range(len(conjunto)):
    for j in range(len(conjunto)):
        if i != j:
            prod = prod_interior(conjunto[i], conjunto[j])
            print(f"<{etiquetas[i]},{etiquetas[j]}> = {prod}")
```

```python
import numpy as np

# Función para calcular productor interior
def prod_interior(a, b):
    result = a[0]*b[0] + a[1]*b[1]
    return result

# Conjunto de datos:
A = [1,0]
B = [0,1]
C = [1,1]
D = [1,-1]
E = [np.sqrt(2)/2, np.sqrt(2)/2]
F = [np.sqrt(2), -np.sqrt(2)]
G = [1,2]
H = [4,-2]

conjunto = [A,B,C,D,E,F,G,H]
etiquetas = ['A','B','C','D','E','F','G','H']

# Cálculo de normas:
for i in range(len(conjunto)):
    prod = np.sqrt(prod_interior(conjunto[i], conjunto[i]))
    print(f"||{etiquetas[i]}|| = {prod}")
```
