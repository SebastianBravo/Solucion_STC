###**2.** Grafique la potencia instantánea y calcule la energía total de las siguientes señales:

####**a.** $a\left(t\right)=A_0\cdot \sin\left(2\pi\cdot1000t\right)$

**Energía total:** $E_\infty\{a(t)\} = \infty$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -0.003

# Tiempo final de la gráfica
t_final = 0.003

# Número de puntos a graficar
puntos = 10000

# Amplitud de la señal
A0 = 2 

# No modificar
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# No modificar
# Para cada valor del vector tiempo, calcula a
a = [A0*x for x in np.sin(2*np.pi*1000*t)]

# Para cada valor de a, calcula la pontencia instantanea
p = [abs(x)**2 for x in a]

# Gráfica de la señal
plt.plot(t,a)
plt.xlabel('t')
plt.ylabel(r'$a(t)$')
plt.title(r'Señal $a(t)$')
plt.grid(True)
plt.show()

# Gráfica de la potencia instantane
plt.plot(t,p)
plt.xlabel('t')
plt.ylabel(r'$P\{a(t)\}$')
plt.title(r'Señal $P\{a(t)\}$')
plt.grid(True)
plt.show()
```

####**b.** $b(t)=1+2\cos(2π\cdot2t)$

####**d.** $d(t) = 
\begin{cases} 
  0 & \text{$t>0$} \\ 
  2\cdot e^{3t} & \text{$t\leq0$}
\end{cases}$

**Energía total:** $E_\infty\{d(t)\} = \frac{2}{3}$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -2

# Tiempo final de la gráfica
t_final = 2

# Número de puntos a graficar
puntos = 10000

# No modificar
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# No modificar
# Para cada valor del vector tiempo, calcula d
d = []
for x in t:
  if (x<=0):
    d.append(2*np.exp(3*x))
  else:
    d.append(0)

# Para cada valor de d, calcula la pontencia instantanea
p = [abs(x)**2 for x in d]

# Gráfica de la señal
plt.plot(t,d)
plt.xlabel('t')
plt.ylabel(r'$d(t)$')
plt.title(r'Señal $d(t)$')
plt.grid(True)
plt.show()

# Gráfica de la potencia instantane
plt.plot(t,p)
plt.xlabel('t')
plt.ylabel(r'$P\{d(t)\}$')
plt.title(r'Señal $P\{d(t)\}$')
plt.grid(True)
plt.show()
```

####**g.** $g(t)=4⋅e^{j2t}$

**Energía total:** $E_\infty\{g(t)\} = \infty$

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -6

# Tiempo final de la gráfica
t_final = 6

# Número de puntos a graficar
puntos = 10000

# Frecuencia angular
w = 2

# Fase
fase = 0

# Amplitud
A = 4

# No modificar a partir de este punto
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Para cada valor en t, calcula la parte real de la señal
sig_real = A*np.cos(w*t - fase)

# Para cada valor en t, calcula la parte imaginaria de la señal
sig_img = A*np.sin(w*t - fase)

# Para cada valor de f, calcula la pontencia instantanea
p = [4**2 for x in t]

# Gráfica de la señal
ax = plt.axes(projection="3d")
ax.plot3D(t,sig_real, sig_img)
ax.set_xlabel('t')
ax.set_ylabel(r'$Real\{g(t)\}$')
ax.set_zlabel(r'$Img\{g(t)\}$')
ax.set_title(r'Señal $g(t)$')
plt.grid(True)
plt.show()

# Gráfica de la parte real de la señal
plt.plot(t,sig_real)
plt.xlabel('t')
plt.ylabel(r'$Real\{g(t)\}$')
plt.title(r'Parte real de $g(t)$')
plt.grid(True)
plt.show()

# Gráfica de la parte imaginaria de la señal
plt.plot(t,sig_img)
plt.xlabel('t')
plt.ylabel(r'$Img\{g(t)\}$')
plt.title(r'Parte imaginaria de $g(t)$')
plt.grid(True)
plt.show()

# Gráfica de la potencia instantane
plt.plot(t,p)
plt.xlabel('t')
plt.ylabel(r'$P\{g(t)\}$')
plt.title(r'Señal $P\{g(t)\}$')
plt.grid(True)
plt.show()
```
