###**4.** Hallar la serie de Fourier de las siguientes señales, aplicando las propiedades mencionadas en el numeral 3:



####**a.** $$z_{1}( t) =7\sin\left(\frac{3\pi }{4} \cdot t\right) +2\cos\left(\frac{3\pi }{4} \cdot t\right)$$





**Señales a usar:** $y_1(t)$, $y_2(t)$

**Propiedades:** Linealidad y escalamiento en el tiempo

$$ \begin{array}{l}
y_{1}( t) =\sum _{k=-\infty }^{\infty } a_{k} \cdot e^{jk5\pi t}\\
\\
a_{1} =a_{-1} =2,\ a_{k} =0\ \forall \ k\neq \pm 1\\
\\
y_{2}( t) =\sum _{k=-\infty }^{\infty } b_{k} \cdot e^{jk3\pi t}\\
\\
b_{1} =\frac{5}{2j} ,b_{-1} =-\frac{5}{2j} ,\ b_{k} =0\ \forall \ k\neq \pm 1\\
\\
z_{1}( t) =\frac{1}{2} \cdot y_{1}\left(\frac{3}{20} \cdot t\right) +\frac{7}{5} \cdot y_{2}\left(\frac{1}{4} \cdot t\right)\\
\\
z_{1}( t) =\sum _{k=-\infty }^{\infty } a_{k} \cdot e^{jk\frac{3\pi }{4} t} +\sum _{k=-\infty }^{\infty } b_{k} \cdot e^{jk\frac{3\pi }{4} t}\\

\end{array}$$
