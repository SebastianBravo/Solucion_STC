###**2.** Hallar el soporte y la cota (si la tienen) de las siguientes señales:

####**a.** $n(t) = 3 - |t|$

**Soporte:** $(-\infty, \infty)$

**Cota:** No posee

**Valor máximo de $n(t)$ :** $3$

**Valor máximo de $|n(t)|$ :** $\infty$

**Represntación gráfica usando python:**


```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -10

# Tiempo final de la gráfica
t_final = 10

# Número de puntos a graficar
puntos = 10000

# No modificar a partir de este punto
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Para cada valor del vector tiempo, calcula n
n = [3 - abs(x) for x in t]

# Para cada valor de n, calcula su valor absoluto
n_abs = [abs(x) for x in n]

# Calcula el valor máximo de la señal y del valor absoluto de la señal
print(f"El valor máximo de n(t) en la gráfica actual es {max(n)}")
print(f"El valor máximo de |n(t)| en la gráfica actual es {max(n_abs)}")

# Gráfica de la señal
plt.plot(t,n)
plt.xlabel('t')
plt.ylabel('n(t)')
plt.title('Señal n(t)')
plt.grid(True)
plt.show()

# Gráfica del valor absoluto de la señal
plt.plot(t,n_abs)
plt.xlabel('t')
plt.ylabel('|n(t)|')
plt.title('Señal |n(t)|')
plt.grid(True)
plt.show()
```

####**d.** $g(t) = 
\begin{cases} 
  0 & \text{$t<0$} \\ 
  e^{-2t} & \text{$t\geq0$}
\end{cases}$

**Soporte:** $[0, \infty)$

**Cota:** 1

**Valor máximo de $g(t)$ :** $1$

**Valor máximo de $|g(t)|$ :** $1$

**Represntación gráfica usando python:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -3

# Tiempo final de la gráfica
t_final = 8

# Número de puntos a graficar
puntos = 10000

# No modificar a partir de este punto
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Para cada valor del vector tiempo, calcula g
g = []
for x in t:
  if x < 0:
    g.append(0)
  else:
    g.append(np.exp(-2*x))

# Para cada valor de g, calcula su valor absoluto
g_abs = [abs(x) for x in g]

# Calcula el valor máximo de la señal y del valor absoluto de la señal
print(f"El valor máximo de g(t) en la gráfica actual es {max(g)}")
print(f"El valor máximo de |g(t)| en la gráfica actual es {max(g_abs)}")

# Gráfica de la Señal
plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('g(t)')
plt.title('Señal g(t)')
plt.grid(True)
plt.show()

# Gráfica del valor absoluto de la Señal
plt.plot(t,g_abs)
plt.xlabel('t')
plt.ylabel('|g(t)|')
plt.title('Señal |g(t)|')
plt.grid(True)
plt.show()
```

####**f.** $w(t)=
\begin{cases} 
  -2t & \text{$1<|t|<3$} \\ 
  0 & \text{En el resto}
\end{cases}$

**Soporte:** $(-3,3)$

**Cota:** *6*

**Valor máximo de $w(t)$ :** $6$

**Valor máximo de $|w(t)|$ :** $6$

**Represntación gráfica usando python:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Tiempo inicial de la gráfica
t_inicial = -5

# Tiempo final de la gráfica
t_final = 5

# Número de puntos a graficar
puntos = 10000

# No modificar a partir de este punto
# Vector tiempo
t = np.linspace(t_inicial,t_final,puntos)

# Para cada valor del vector tiempo, calcula w
w = []
for x in t:
  if (abs(x)<3) and (abs(x)>1):
    w.append(-2*x)
  else:
    w.append(0)

# Para cada valor de w, calcula su valor absoluto
w_abs = [abs(x) for x in w]

# Calcula el valor máximo de la señal y del valor absoluto de la señal
print(f"El valor máximo de w(t) en la gráfica actual es {max(w)}")
print(f"El valor máximo de |w(t)| en la gráfica actual es {max(w_abs)}")

# Gráfica de la Señal
plt.plot(t,w)
plt.xlabel('t')
plt.ylabel('w(t)')
plt.title('Señal w(t)')
plt.grid(True)
plt.show()

# Gráfica del valor absoluto de la Señal
plt.plot(t,w_abs)
plt.xlabel('t')
plt.ylabel('|w(t)|')
plt.title('Señal |w(t)|')
plt.grid(True)
plt.show()
```
