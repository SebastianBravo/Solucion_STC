**Cota:** No posee

####**a.** $a(t) = 
\begin{cases} 
  t & \text{$|t|<4$} \\ 
  0 & \text{$|t|\geq4$}
\end{cases}$

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector tiempo
t = np.linspace(-6,6,1000)
a = []
a_par = []
a_impar =[]

# Evalua cada tiempo en la señal
for x in t:
  if (abs(x)<4):
    a.append(x)
  else:
    a.append(0)

# Invierte la señal en tiempo
a_inv = list(reversed(a))

# Calcula la parte par e impar de la señal
for i in range(len(a)):
  a_par.append((a[i] + a_inv[i])/2)
  a_impar.append((a[i] - a_inv[i])/2)

# Gráfica de la señal
plt.plot(t,a)
plt.xlabel('t')
plt.ylabel('a(t)')
plt.title('Señal a(t)')
plt.grid(True)
plt.show()

# Gráfica de la componente par de la señal
plt.plot(t,a_par, label='Parte par')
plt.xlabel('t')
plt.title('Componente par de a(t)')
plt.ylim(-4, 4)  
plt.grid(True)
plt.show()

# Gráfica de la componente impar de la señal
plt.plot(t,a_impar, label='Parte impar')
plt.xlabel('t')
plt.title('Componente impar de a(t)')
plt.grid(True)
plt.show()
```

###**1.** Graficar las siguientes señales. Indique tiempos y amplitudes relevantes: 

####**a.** $2u(t)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector tiempo
t = np.linspace(-6,6,1000)

def u(t, a, b, t0, c):
  """
  Función que crea señal au(bt + t0) + c 
  Toma como parametros un vector de tiempo(t) y a, b, t0 y c
  Retorna una pareja de tiempo y amplitud
  """
  amp = []
  
  # Crea funcion u(t)
  for x in t:
    if x >= 0:
      amp.append(1)
    else:
      amp.append(0)

  # Desplazamiento y escalado en tiempo
  t = [(x+t0)/b for x in t]

  # Desplazamiento en eje vertical y escalado en amplitud
  amp = [(i*a)+c for i in amp]

  return (t, amp)

# Define la señal
t, sig = u(t, 2, 1, 0, 0)

# Gráfica de la señal
plt.plot(t,sig)
plt.xlabel('t')
plt.ylabel(r'$2u(t)$')
plt.title(r'Señal $2u(t)$')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector tiempo
t = np.linspace(-2,2,1000000)

def dirac_delta(t, a, b, t0, c):
  """
  Función que crea señal au(bt + t0) + c 
  Toma como parametros un vector de tiempo(t) y a, b, t0 y c
  Retorna una pareja de tiempo y amplitud
  """
  # Entre más pequeño sea delta, mejor es la representación del impulso
  delta = 1/1000
  amp = []
  
  # Crea funcion dirac_delta(t)
  for x in t:
    if (x >= -delta/2) and (x <= delta/2):
      amp.append(1/delta)
    else:
      amp.append(0)

  # Desplazamiento y escalado en tiempo
  t = [(x+t0)/b for x in t]

  # Desplazamiento en eje vertical y escalado en amplitud
  amp = [(i*a)+c for i in amp]

  return (t, amp)

def integral(t, amp):
  """
  Función que calcula la integral de una señal
  Toma como parametros una pareja de (tiempo, amplitud)
  Retorna el resultado de la integral
  """
  dt = (max(t)- min(t))/len(t)
  area = 0

  for a in amp:
    area += dt*a

  return area

# Define la señal
t, sig = dirac_delta(t, 1, 1, 2, 0)

# Imprime el area bajo la curva de la señal
print(f"El área bajo la curva de la señal es: {integral(t, sig)}")

# Gráfica de la señal
plt.plot(t,sig)
plt.xlabel('t')
plt.ylabel(r'$\delta (t-2)$')
plt.title(r'Señal $\delta (t-2)$')
plt.grid(True)
plt.show()
```
